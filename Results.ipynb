<h1>American Sign Language - Object Detection using YOLOv5</h1>
<h3>In this notebook, the model architecture, dataset used, and the results are presented.</h3> 

<h2>Dataset</h2>

<h5>The American Sign Language Letters dataset is an object detection dataset of each ASL letter with a bounding box. David Lee, a data scientist focused on accessibility, curated and released the dataset for public use.</h5>

[Link to the Dataset ASL](https://public.roboflow.com/object-detection/american-sign-language-letters)

<p>This dataset contains 26 classes for each letter: 'A', 'B', 'C', ..., 'Z'</p>

<ul>
    <li>Number of Images: 1728</li>
    <li>Image size: 416x416</li>
    <li>Number of images in Training set: 1512</li>
    <li>Number of images in Validation set: 144</li>
    <li>Number of images in Test set: 72</li>
</ul>


<h1> Model Selection </h1>

<p> I chose to use the YOLOv5s model for object detection</br>

I trained the YOLOv5s model on the dataset using the following default configuration:</br>

Batch Size: 1</br>
Number of Epochs: 50 and 100</br>
Initial Learning Rate: 0.01</br>
Optimizer: : SGD with momentum (0.937) and weight decay (0.0005)</br>
Loss Function: BCE (Binary Cross Entropy) with logits</br>
</p>

<h1>Development Environemnt</h1>

The development Environment was setup according to the Yolov5 repository requirements.txt</br>

I Used:
- **Python Version:** 3.9.16
- **PyTorch Version:** 1.11.0 + CUDA Version 11.3

<h1> Training Results <h2>

<h4>1. When comparing the training results on validation set , the Precision, Recall and mAP50 scores improved when the model was trained for 100 epochs as compared to 50 epochs</h4>

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to your images
image_path1 = "50epochsvalid.png"
image_path2 = "100epochvalid.png"

# Load the images
img1 = mpimg.imread(image_path1)
img2 = mpimg.imread(image_path2)

# Set the size of the plot
plt.figure(figsize=(20, 10))  # Adjust the size as needed


# Display first image
plt.subplot(2, 1, 1)
plt.imshow(img1)
plt.axis('off')  # Hide axis
plt.title('Model trained with 50 epochs')  # Add caption

# Display second image
plt.subplot(2, 1, 2)
plt.imshow(img2)
plt.axis('off')  # Hide axis
plt.title('Model trained with 100 epochs')  # Add caption

plt.show()

```

<h4>2. Convergence: </h4>
<h5>The model converges better in case of 100 epochs than 50 epochs. The training and validation bounding box regression loss, objs_loss(the confidence of object presence is the objectness loss ) stops decreasing indicating convergence.</h5>

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Path to your images
image_path1 = "yolov5s_resultsEpoch50//results.png"
image_path2 = "yolov5s_results2Epoch100//results.png"

# Load the images
img1 = mpimg.imread(image_path1)
img2 = mpimg.imread(image_path2)

# Set the size of the plot
plt.figure(figsize=(20, 20))  # Adjust the size as needed


# Display first image
plt.subplot(2, 1, 1)
plt.imshow(img1)
plt.axis('off')  # Hide axis
plt.title('Model trained with 50 epochs')  # Add caption

# Display second image
plt.subplot(2, 1, 2)
plt.imshow(img2)
plt.axis('off')  # Hide axis
plt.title('Model trained with 100 epochs')  # Add caption

plt.show()
```
